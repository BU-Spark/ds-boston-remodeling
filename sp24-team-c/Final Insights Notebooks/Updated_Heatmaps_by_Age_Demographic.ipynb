# Neighborhoods by Age Demographics

Okay so based on the bedrooms/res_units (doesn't matter too much which one we use) maps we can now map age demographic data onto it to see which demographics are being affected by the changes. First lets get the neighborhood data previous notebooks and load in the age demographic data.

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
import numpy as np


bedrooms_neighborhoods_df = pd.read_csv('../Demographic Data/bedrooms_neighborhoods_gdf.csv')

# Convert the 'geometry' column from WKT string back to geometry objects
bedrooms_neighborhoods_df['geometry'] = bedrooms_neighborhoods_df['geometry'].apply(wkt.loads)

# Convert the pandas DataFrames into GeoDataFrames
bedrooms_neighborhoods_df = gpd.GeoDataFrame(bedrooms_neighborhoods_df, geometry='geometry')
```

```python
display(bedrooms_neighborhoods_df.head())
```

```python
# Set the CRS for the GeoDataFrames to the known CRS
bedrooms_neighborhoods_df.crs = 'EPSG:4326'
```

```python
age_demographic_data = pd.read_csv('../Demographic Data/Age_Demographics.csv')
#household_demographic_data = pd.read_csv('../Demographic Data/Household_Demographics.csv')
```

```python
display(age_demographic_data.head())
```

```python
# Checking the data types of the columns
age_demographic_data.dtypes
```

```python
# Remove leading and trailing spaces from column names
age_demographic_data.columns = age_demographic_data.columns.str.strip()
```

```python
# Checking the data types of the columns
age_demographic_data.dtypes
```

```python
# Convert 'Total Population' and any other population columns to numeric, removing any commas first
age_demographic_data['Total Population'] = pd.to_numeric(age_demographic_data['Total Population'].str.replace(',', ''), errors='coerce')

# Now do the same for any other population columns that are stored as objects
pop_columns = ['0-9 years Pop', '10-17 years Pop', '18-19 years Pop', '20-34 years Pop', '35-59 years Pop', '60+ years Pop']
for col in pop_columns:
    age_demographic_data[col] = pd.to_numeric(age_demographic_data[col].str.replace(',', ''), errors='coerce')

# Confirm the changes
age_demographic_data.dtypes
```

```python
percent_columns = ['0-9 Percent', '10-17  Percent', '18-19 Percent', '20-34 Percent', '35-59 Percent', '60+ Percent']

for column in percent_columns:
    age_demographic_data[column] = pd.to_numeric(age_demographic_data[column].str.replace('%', '').str.strip(), errors='coerce')

# Confirm the changes
print(age_demographic_data.dtypes)
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd

# Strip whitespace and convert to consistent case (e.g., uppercase) in the demographic DataFrame
age_demographic_data['Neighborhood'] = age_demographic_data['Neighborhood'].str.strip().str.upper()

# Do the same for the neighborhoods GeoDataFrame
bedrooms_neighborhoods_df['BlockGr202'] = bedrooms_neighborhoods_df['BlockGr202'].str.strip().str.upper()


# Find the modal age group by taking the idxmax of the percentage columns along the rows
age_columns = [col for col in age_demographic_data.columns if 'Percent' in col]
age_demographic_data['Modal_Age_Group'] = age_demographic_data[age_columns].idxmax(axis=1)

# Merge the modal age group data with the neighborhood GeoDataFrame
# Make sure the neighborhood names match between the two DataFrames and are in the same case
# Make sure the merge is being done in the right order and with the right column names
bedrooms_neighborhoods_df = bedrooms_neighborhoods_df.merge(age_demographic_data[['Neighborhood', 'Modal_Age_Group']], left_on='BlockGr202', right_on='Neighborhood', how='left')

```

```python
display(bedrooms_neighborhoods_df)
```

```python


import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from shapely.geometry import Polygon

# Create a color map for the modal age group
color_map = {
    '0-9 Percent': 'yellow',
    '10-17  Percent': 'orange',
    '18-19 Percent': 'red',
    '20-34 Percent': 'green',
    '35-59 Percent': 'blue',
    '60+ Percent': 'purple',
    'NaN': 'grey'  # Assign a default color for NaN values
}

# Map the modal age group to its corresponding color
bedrooms_neighborhoods_df['color'] = bedrooms_neighborhoods_df['Modal_Age_Group'].map(color_map)

# Replace NaN values in 'color' with the default color for NaN values
bedrooms_neighborhoods_df['color'].fillna(color_map['NaN'], inplace=True)

# Plot the map
fig, ax = plt.subplots(1, figsize=(15, 15))
bedrooms_neighborhoods_df.plot(ax=ax, color=bedrooms_neighborhoods_df['color'], edgecolor='black', linewidth=0.5)

# Create a patch (proxy artist) for each color map
patches = [mpatches.Patch(color=color, label=age_group) for age_group, color in color_map.items() if age_group != 'NaN']
plt.legend(handles=patches, title='Age Groups', loc='upper left', fontsize=12)

# Annotate the neighborhoods with their names
for idx, row in bedrooms_neighborhoods_df.iterrows():
    # Skip rows where the geometry is empty or invalid
    if isinstance(row['geometry'], Polygon):
        plt.annotate(text=row['Neighborhood'], 
                     xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                     horizontalalignment='center', verticalalignment='center', fontsize=8)

# Remove axis off
ax.axis('off')

# Add title
plt.title('Modal Age Group per Neighborhood', fontsize=20, y=1.02)


# Show the map
plt.show()

```

```python
import matplotlib.pyplot as plt


# Define the categories and colors for the pie charts
age_categories = ['0-9 Percent', '10-17  Percent', '18-19 Percent', 
                  '20-34 Percent', '35-59 Percent', '60+ Percent']
colors = ['yellow', 'orange', 'red', 'green', 'blue', 'purple']

# Determine the layout of the subplots
nrows = int(np.ceil(len(bedrooms_neighborhoods_df) / 3.0))
ncols = 3

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, nrows * 7))

# Flatten the axes array for easy iteration
axes_flat = axes.flatten()

# Plot a pie chart for each neighborhood
for idx, (ax, neighborhood) in enumerate(zip(axes_flat, bedrooms_neighborhoods_df['Neighborhood'])):
    # Get the age distribution for the current neighborhood
    distribution = age_demographic_data[age_demographic_data['Neighborhood'] == neighborhood][age_categories].values.flatten().tolist()
    
    # If the neighborhood has no data, continue to the next iteration
    if not distribution:
        ax.axis('off')  # Hide the axis if no data
        continue
    
    # Create the pie chart
    ax.pie(distribution, labels=age_categories, colors=colors, autopct='%1.1f%%', startangle=140)
    ax.set_title(neighborhood)

# Hide any unused axes
for ax in axes_flat[idx+1:]:
    ax.axis('off')

# Add an overall title
plt.suptitle('Age Demographics by Neighborhood', fontsize=16, y=1.02)

# Adjust layout to prevent overlap, you can fine-tune these parameters
# plt.subplots_adjust(hspace=0.2, wspace=0.5)



# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
