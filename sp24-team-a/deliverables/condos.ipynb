```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have data for each year in separate CSV files named FY01.csv, FY02.csv, ..., FY20.csv
# You can adjust the file names and paths according to your actual data structure.

# Initialize an empty dictionary to store counts for each year
condo_counts = {}

# Loop through each year
for year in range(21, 25):
    # Read the CSV file for the current year into a DataFrame
    filename = f'FY{year:d}.csv'
    df = pd.read_csv(filename)
    
    # Count the occurrences of 'CONDO' in the 'LU_DESC' column
    count = df['LU_DESC'].str.contains('CONDO', case=False).sum()
    
    # Store the count for the current year
    condo_counts[year] = count

# Convert the dictionary to a pandas Series for easier plotting
condo_counts_series = pd.Series(condo_counts)

# Plot the counts over the years
condo_counts_series.plot(kind='bar', figsize=(10, 6))
plt.title('Count of CONDO Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Read data for FY14 and FY24
fy14 = 'FY14.csv'
fy14df = pd.read_csv(fy14)
fy24 = 'FY24.csv'
fy24df = pd.read_csv(fy24)

column_name = 'ZIPCODE'

# Remove leading zeros and underscores from the zip code column
fy14df[column_name] = fy14df[column_name].str.lstrip('0')  # Remove leading zeros
fy14df[column_name] = fy14df[column_name].str.rstrip('_')         # Remove trailing underscore

fy14df = fy14df.dropna(subset=['LU'])

fy24df = fy24df.rename(columns={'ZIP_CODE':'ZIPCODE'})

# Filter data for condos
fy14_condo = fy14df[fy14df['LU'].str.contains('CD', case=False)].groupby('ZIPCODE').size().rename('FY14')
print(fy14_condo)
fy24_condo = fy24df[fy24df['LU_DESC'].str.contains('CONDO', case=False)].groupby('ZIPCODE').size().rename('FY24')
print(fy24_condo)

condo_comparison = pd.merge(fy14_condo, fy24_condo, how='outer', left_index=True, right_index=True)

# Reset the index
condo_comparison.reset_index(inplace=True)

# Plotting
plt.figure(figsize=(15, 8))
for zipcode in condo_comparison['ZIPCODE'].unique():
    data = condo_comparison[condo_comparison['ZIPCODE'] == zipcode]
    plt.plot(['FY14', 'FY24'], data[['FY14', 'FY24']].values.flatten(), label=zipcode)

plt.title('Comparison of CONDO Counts by Zip Code between FY14 and FY24')
plt.xlabel('Year')
plt.ylabel('Count of CONDO')
plt.xticks(rotation=45)
plt.legend(title='Zip Code', loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
