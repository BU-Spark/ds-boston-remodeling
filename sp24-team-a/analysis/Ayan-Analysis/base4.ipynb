```python
vals = {}

for i in range(2014, 2025):
    f = f'PropertyAssessmentDatasets/assess{i}.csv'
    df = pd.read_csv(f)
    exists = 'LU' in df.columns
    vals[str(i)] = exists

print(vals)
```

```python
vals = {}

for i in range(2014, 2025):
    f = f'PropertyAssessmentDatasets/assess{i}.csv'
    df = pd.read_csv(f)
    count = 0

    lu_counts = df['LU'].value_counts()

    for j in range(1, 5):
        if f'R{j}' in lu_counts:
            count += j * lu_counts[f'R{j}']
    
    vals[i] = count

print(vals)
```

```python
import matplotlib.pyplot as plt

# Sorting the dictionary by year (key) and unpacking into lists
years, values = zip(*sorted(vals.items()))

# Creating the plot
plt.figure(figsize=(10, 6))  
plt.plot(years, values, marker='o', linestyle='-', color='b') 

# Adding title and labels
plt.title('Change in Residential Units Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Units')


plt.grid(True)

# Display the plot
plt.show()
```

```python
sorted_data = sorted(vals.items())

# Loop through the sorted items and calculate the change for each year
for i in range(1, len(sorted_data)):
    # Current year and value
    year, value = sorted_data[i]
    # Previous year and value
    prev_year, prev_value = sorted_data[i-1]
    
    # Calculate the change from the previous year
    change = value - prev_value
    
    # Print the change
    print(f"Change from {prev_year} to {year}: {change} units")
```
