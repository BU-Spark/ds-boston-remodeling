```python
import pandas as pd
```

```python
approved_permits = pd.read_csv('approvedpermits.csv')
approved_permits.head()
```

```python
#Cleaning stuff - Still have more to clean. Seems more manually though
approved_permits['city'] = approved_permits['city'].str.lower()
# approved_permits['city'] = approved_permits['city'].str.replace(' ', '', regex=False)
approved_permits['city'] = approved_permits['city'].str.replace('/boston', '', regex=False)
approved_permits['city'] = approved_permits['city'].str.replace('boston/', '', regex=False)
approved_permits['city'] = approved_permits['city'].str.replace(r'[\!\"\#\$\%\&\*\+\,\-\.\/\:;\<\=\>\?\@\[\\\]\^\_\`\{\|\}\~\']+$', '', regex=True)

approved_permits['declared_valuation'] = approved_permits['declared_valuation'].str.replace(r'[\$,]', '', regex=True)
approved_permits['declared_valuation'] = approved_permits['declared_valuation'].astype(float)


```

```python
# Find relationship between city and most common worktype/permit type descr
# Find relationship between city and most common issued_date
# Find relationship between city and declared valuation
#["city","issued_date","declared_valuation","permittypedescr", "worktype"]
```

```python
common_vals_city = approved_permits[["city","worktype","issued_date","permittypedescr"]].groupby("city").apply(lambda x: x.mode().iloc[0])
common_vals_city.head(55)
```

```python
valuation_city = approved_permits[["city","declared_valuation"]].groupby("city").agg("mean")
pd.set_option('display.float_format', '{:.2f}'.format)

valuation_city.head(55)
```

```python
#2024
#Importing the dataset
df = pd.read_csv('PropertyAssessmentDatasets/fy2024-property-assessment-data_1_5_2024.csv')

#Cleaning:
df["GROSS_AREA"] = pd.to_numeric(df["GROSS_AREA"], errors='coerce')
df = df.dropna(subset=["GROSS_AREA"])

#Create Groupby
area_city2024 = df[["CITY","GROSS_AREA"]].groupby("CITY").agg("mean")
area_city2024.head()
```
